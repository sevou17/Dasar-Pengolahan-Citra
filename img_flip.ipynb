```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar
img = cv2.imread('kadal.jpg')

# Flip gambar
flipped_img_horizontal = cv2.flip(img, 1)  # flip horizontal
flipped_img_vertical = cv2.flip(img, 0)    # flip vertikal

# Menampilkan gambar asli dan hasil flip
plt.figure(figsize=(15,10 ))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis()

plt.subplot(1, 3, 2)
plt.title('Horizontally Flipped Image')
plt.imshow(cv2.cvtColor(flipped_img_horizontal, cv2.COLOR_BGR2RGB))
plt.axis()

plt.subplot(1, 3, 3)
plt.title('Vertically Flipped Image')
plt.imshow(cv2.cvtColor(flipped_img_vertical, cv2.COLOR_BGR2RGB))
plt.axis()

plt.show()
```
